```python
from glob import glob
import os
import pickle as pkl
import numpy as np
```

# Results of optimizing then testing

```python
os.chdir("/Users/djpassey/Data/RCAlgorithm/results-0/")
results=[]
for path in glob("**/*.pkl", recursive=True):
    with open(path,'rb') as file:
        d=pkl.load(file)
        d["filename"]=path
        results.append(d)
```

```python
results[0:3]
```

```python
runresults = [
    (r["filename"], np.array(r["cont_deriv_fit"]), np.array(r["rand_deriv_fit"]))
    for r in results if "continue" in r.keys()
]
```

```python
for name in ["thomas", "rossler", "lorenz"]:
    for rr in runresults:
        if rr[0].startswith(name):
            print(rr[0])
            print(f"\tContinued Prediction Error: \n\t\t{np.mean(rr[1], axis=0)} (true system, reservoir)")
            print(f"\tRandom Initial Prediction Error: \n\t\t{np.mean(rr[1], axis=0)} (true system, reservoir)")
            print("\n")
```
