# Nearby Initial Value Test

```python
import rescomp as rc
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [10,10]
```

```python
PRMS = {
    "res_sz": 1000,
    "activ_f": np.tanh,
    "mean_degree": 2.0,
    "ridge_alpha": 1e-6,
    "spect_rad": 0.9,
    "gamma": 5.,
    "sigma": 0.14,
    "uniform_weights": True,
    "sparse_res": True,
    "signal_dim": 3
}
```

```python
t, U = rc.orbit("lorenz", duration=50, trim=True)
rcomp = rc.ResComp(**PRMS)
rcomp.train(t, U)
u0 = U[0, :] + np.random.randn(3) * 1e-4
pre = rcomp.predict(t, u0)
rc.plot3d(pre)
```

```python
t, U = rc.orbit("lorenz", duration=50, trim=True)
rcomp = rc.ResComp(**PRMS)
rcomp.train(t, U, window=10, overlap=0.95)
u0 = U[0, :] + np.random.randn(3) * 1e-4
pre = rcomp.predict(t, u0)
rc.plot3d(pre)
```

## Random Initial on Attractor

```python
t, U = rc.orbit("lorenz", duration=50, trim=True)
rcomp = rc.ResComp(**PRMS)
rcomp.train(t, U)
ts, Uts = rc.orbit("lorenz", duration=1000, trim=True)
u0 = Uts[0, :]
pre = rcomp.predict(ts, u0)
rc.plot3d(pre)
```

```python
t, U = rc.orbit("lorenz", duration=50, trim=True)
rcomp = rc.ResComp(**PRMS)
rcomp.train(t, U, window=10.0, overlap=0.5)
ts, Uts = rc.orbit("lorenz", duration=1000, trim=True)
u0 = Uts[0, :]
pre = rcomp.predict(ts, u0)
rc.plot3d(pre)
```

```python
t, U = rc.orbit("lorenz", duration=50, trim=True)
rcomp = rc.ResComp(**PRMS)
rcomp.train(t, U, window=10.0, overlap=.95)
ts, Uts = rc.orbit("lorenz", duration=1000, trim=True)
u0 = Uts[0, :]
pre = rcomp.predict(ts, u0)
rc.plot3d(pre)
```
